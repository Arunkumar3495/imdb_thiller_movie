```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
url = 'https://www.imdb.com/search/title/?genres=Thriller&sort=user_rating,desc&title_type=feature&num_votes=25000,&ref_=chtmvm_gnr_20'

response = requests.get(url)

print(response)
```

```python
import re
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text,'html.parser')
```

```python
listings = soup.find('div',class_='lister-list').find_all('div',class_='lister-item mode-advanced')

print(len(listings))
```

```python
thiller_movie_list = []
for movie in listings:
  index = movie.find('h3').find('span',class_='lister-item-index unbold text-primary').get_text().split('.')[0]
  movie_name = movie.find('h3').a.text
  rating = movie.find('div',class_='ratings-bar').strong.text.split('.')[0]
  story = movie.find('p').findNext('p').get_text(strip=True)
  director = movie.find('p').findNext('p').findNext('p').a.text
  gross = movie.find('p',class_='sort-num_votes-visible').find_all('span')[-1].get_text()
  print(index,movie_name,rating,story,director,gross)

  thiller_movie_list.append({
      'Rank':index,
      'Name':movie_name,
      'Rate':rating,
      'Story':story,
      'Director':director,
      'Gross':gross,

  })
```

```python
Thiller = pd.DataFrame(thiller_movie_list)
```
